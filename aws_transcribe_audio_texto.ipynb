```python
import boto3
import time
import urllib
import json
import urllib.request
```

```python
AWS_ACCESS_KEY_ID = '<<ACCESS_KEY_ID>>'
AWS_SECRET_ACCESS_KEY = '<<AWS_SECRET_ACCESS_KEY>>'

job_name = ['AudioAWSYoutubeV2','AudioPalestraYoutubeV2', 'AudioAWSWhatsAppV2']
audios   = ['audio_o_que_e_aws.mp3', 'audio_palestra_cortela.mp3', 'audio_whatsapp_aws.mp3']
```

```python
contador = 0

for i in audios:
  job_uri = f's3://testeaudioawsmolina/{i}'
  Transcribe = boto3.client('transcribe', aws_access_key_id=AWS_ACCESS_KEY_ID, aws_secret_access_key=AWS_SECRET_ACCESS_KEY, region_name='sa-east-1')
  Transcribe.start_transcription_job(TranscriptionJobName=job_name[contador], Media={'MediaFileUri': job_uri}, MediaFormat='mp3', LanguageCode='pt-BR')

  while True:
    status = Transcribe.get_transcription_job(TranscriptionJobName=job_name[contador])
    if status['TranscriptionJob']['TranscriptionJobStatus'] in ['COMPLETED', 'FAILED']:
        break
    time.sleep(2)

  if status['TranscriptionJob']['TranscriptionJobStatus'] == 'COMPLETED':
    with urllib.request.urlopen(status['TranscriptionJob']['Transcript']['TranscriptFileUri']) as url:
      response = url.read()
      data = json.loads(response)
      text = data['results']['transcripts'][0]['transcript']
      print(job_name[contador] + ': ' + text)

  contador = contador + 1
```
